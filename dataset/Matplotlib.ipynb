```python
z = 1 / (1+np.exp(-x))
plt.plot(x,z,
         color='red',
         lw =3,
         ls= '--',
         marker='o',
         markersize=10,
         markerfacecolor='yellow',
         markeredgecolor='blue',  
         markeredgewidth=4, 
         alpha=0.2
        )

```

```python
z = 1 / (1+np.exp(-x))
plt.plot(x,z,label = 'sigmoid')
plt.plot(x,z*2,label='sigmoid*2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc = 4)
plt.xticks([1,2,3],rotation=20)
```
